```python
users = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv')
sessions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv')
products = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv')
transactions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')
```

```python
#12. Join users to transactions, keeping all rows from transactions and only matching rows from users (left join)
pd.merge(transactions, users, how='left',on=['UserID'])
```

```python
#13. Which transactions have a UserID not in users?
user_uid_list = list(users.UserID)
transaction_uid_list = list(transactions.UserID)
only_transaction_uid_list = list()
for item in transaction_uid_list:
    if item in user_uid_list:
        only_transaction_uid_list.append(False)
    else:
        only_transaction_uid_list.append(True)
transactions.loc[only_transaction_uid_list,:].head()
```

```python
#14. Join users to transactions, keeping only rows from transactions and users that match via UserID (inner join)
pd.merge(transactions, users, how='inner',on=['UserID'])
```

```python
#15. Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)
pd.merge(transactions, users, how='outer',on=['UserID'])
```

```python
#16. Determine which sessions occurred on the same day each user registered
pysqldf("SELECT users.*,sessions.SessionID,sessions.SessionDate from users, sessions where users.UserID=sessions.UserID and users.Registered=sessions.SessionDate and users.UserID in (SELECT users.userID from users, sessions where users.userID = sessions.userID group by users.userID having count(users.userID)==1);")
```

```python
#17. Build a dataset with every possible (UserID, ProductID) pair (cross join)
pysqldf("SELECT distinct users.UserID,transactions.ProductID from users cross join transactions order by users.UserID,transactions.ProductID;")
```

```python
#18. Determine how much quantity of each product was purchased by each user
user_prod_quan = pysqldf("SELECT UserID, ProductID,sum(Quantity) as Quantity from transactions group by UserID, ProductID;")
user_prod = pysqldf("SELECT DISTINCT b.UserID, a.ProductID FROM transactions a, transactions b;")
user_prod['Quantity'] = np.zeros(len(user_prod))
quantity_product_user = pd.merge(user_prod, user_prod_quan, how='left',on=['UserID','ProductID'])
quantity_product_user['Quantity'] = quantity_product_user['Quantity_x'] + quantity_product_user['Quantity_y']
quantity_product_user.drop(['Quantity_x', 'Quantity_y'], axis=1, inplace=True)
quantity_product_user.Quantity.fillna(0, inplace=True)
quantity_product_user.sort_values(['UserID','ProductID'], axis=0, kind='mergesort', inplace=True)
quantity_product_user
```

```python
#19. For each user, get each possible pair of pair transactions (TransactionID1,TransacationID2)
transactions_other = transactions.copy(deep=True)
pair_of_pair_transactions = pd.merge(transactions, transactions_other, how='left',on='UserID')
pair_of_pair_transactions.sort_values('UserID', ascending=False, axis=0, kind='mergesort', inplace=True)
pair_of_pair_transactions
```

```python
#20. Join each user to his/her first occuring transaction in the transactions table
first_trans_per_user = pd.merge(users, transactions, how='left',on='UserID')
first_trans_per_user.drop_duplicates(subset='UserID', keep='first', inplace=True)
first_trans_per_user
```

```python
missing_info = list(first_trans_per_user.columns[first_trans_per_user.isnull().any()])
missing_info
```

```python
for col in missing_info:
    num_missing = first_trans_per_user.loc[first_trans_per_user[col].isnull() == True].shape[0]
    print('number missing for column {}: {}'.format(col, num_missing))  #count of missing data
```

```python
for col in missing_info:
    percent_missing = first_trans_per_user.loc[first_trans_per_user[col].isnull() == True].shape[0] / first_trans_per_user.shape[0]
    print('percent missing for column {}: {}'.format(col, percent_missing))
```
